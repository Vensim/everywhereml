# everywhereml.classification.sklearn.tree.DecisionTreeClassifier

See [sklearn.tree.DecisionTreeClassifier](http://scikit-learn.org/stable/modules/tree.html).

```python
import sys
sys.path.append('../../../../')

from everywhereml.classification.sklearn.tree import DecisionTreeClassifier
from everywhereml.data.plot import scatter
from sklearn.datasets import load_iris
```

```python
"""
Load Iris
"""
X, y = load_iris(return_X_y=True)

scatter(X, hue=y, title='Original')
scatter(X, hue=y, tsne=2, title='Original (t-SNE)')
```

```python
"""
Classify
"""
clf = DecisionTreeClassifier(max_depth=5)
clf.fit(X, y)
y_pred = clf.predict(X)

scatter(X, hue=y_pred, title='Predicted')
scatter(X, hue=y_pred, tsne=2, title='Predicted (t-SNE)')
```

```python
"""
Confusion matrix
"""
clf.plot_confusion_matrix(X, y)
```

```python
"""
Classification errors
"""
scatter(X, hue=(y == y_pred), title='Classification errors')
scatter(X, hue=(y == y_pred), tsne=2, title='Classification errors (t-SNE)')
```
