# everywhereml.data.preprocessing.Window

`Window` applies a rolling window to the input, with or without overlapping.

**it only makes sense with time series data**

```python
import sys
sys.path.append('../../../../')

import numpy as np
from everywhereml.data.preprocessing.time_series import Window
from everywhereml.data.plot import scatter, line
```

```python
"""
Sine series
"""
series = np.sin(np.linspace(-8 * np.pi, 8 * np.pi, 80)).reshape((-1, 1))
line(series, title='Original time series')
```

```python
"""
Apply windowing with overlap
"""
Xt, yt = Window(length=16, overlap=8).fit_transform(series)

print('Xt.shape', Xt.shape)
```

```python
"""
Plot overlapping windows
"""
line(*(Xt.tolist()))
```

```python
"""
Port to C++
"""
print(Window(length=16, overlap=8).fit(series).port(language='cpp'))
```
