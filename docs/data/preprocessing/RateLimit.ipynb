# RateLimit

`RateLimit` artificially creates the desired input frequency. In this case, frequency is considered **in terms of samples, not time!**.

To achieve the desidered frequency it simply drops all but one input every `once_every` input samples.

```python
import sys
sys.path.append('../../../')

from everywhereml.data.preprocessing import RateLimit
from everywhereml.data.plot import scatter
from sklearn.datasets import load_iris
```

```python
"""
Load Iris
"""
X, y = load_iris(return_X_y=True)

print('Number of input samples', len(X))

scatter(X, hue=y, title='Original data')
scatter(X, hue=y, tsne=2, title='Original data (t-SNE)')
```

```python
"""
Xt will contain the diff of each sample from the previous
"""
Xt, yt = RateLimit(once_every=10).fit_transform(X, y)

print('Number of output samples', len(Xt))

scatter(Xt, hue=yt, title='Transformed data')
scatter(Xt, hue=yt, tsne=2, title='Transformed data (t-SNE)')
```

```python
"""
Port to C++
"""
print(RateLimit(once_every=10).fit(X, y).port(language='cpp'))
```

```python
"""
Port to JS
"""
print(RateLimit(once_every=10).fit(X, y).port(language='js'))
```

```python
"""
Port to PHP
"""
print(RateLimit(once_every=10).fit(X, y).port(language='php'))
```
