# everywhereml.data.preprocessing.StandardScaler

See [sklearn.preprocessing.StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html).

```python
import sys
sys.path.append('../../../')

from everywhereml.data.preprocessing import StandardScaler
from everywhereml.data.plot import scatter
from sklearn.datasets import load_iris
```

```python
"""
Load Iris
"""
X, y = load_iris(return_X_y=True)

print('Original range')
print('Mean', X.mean())
print('Std', X.std())

scatter(X, hue=y, title='Original data')
scatter(X, hue=y, tsne=2, title='Original data (t-SNE)')
```

```python
"""
Xt will hold values with zero-mean and unit-variance
"""
Xt, yt = StandardScaler().fit_transform(X, y)

print('Transformed range')
print('Mean', Xt.mean(axis=0))
print('Std', Xt.std(axis=0))

scatter(Xt, hue=yt, title='Transformed data')
scatter(Xt, hue=yt, tsne=2, title='Transformed data (t-SNE)')
```

```python
"""
You can also scale using a global min/max
"""
Xt, yt = StandardScaler(num_features=0).fit_transform(X, y)

print('Transformed range')
print('Min', Xt.mean(axis=0))
print('Max', Xt.std(axis=0))

scatter(Xt, hue=yt)
scatter(Xt, hue=yt, tsne=2)
```

```python
"""
Port to C++
"""
print(StandardScaler().fit(X, y).port(language='cpp'))
```

```python
"""
Port to JS (ES6 dialect)
"""
print(StandardScaler().fit(X, y).port(language='js', dialect='es6'))
```
