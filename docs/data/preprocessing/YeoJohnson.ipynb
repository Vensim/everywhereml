# everywhereml.data.preprocessing.YeoJohnson

See [sklearn.preprocessing.PowerTransformer](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PowerTransformer.html).

See [Yeo-Johnson Wikipedia](https://en.wikipedia.org/wiki/Power_transform#Yeo-Johnson_transformation)

```python
import sys
sys.path.append('../../../')

from everywhereml.data.preprocessing import YeoJohnson
from everywhereml.data.plot import scatter
from sklearn.datasets import load_iris
```

```python
"""
Load Iris
"""
X, y = load_iris(return_X_y=True)

print('Original range')
print('Min', X.min(axis=0))
print('Max', X.max(axis=0))
print('Avg', X.mean(axis=0))
print('Std', X.std(axis=0))

scatter(X, hue=y, title='Original data')
scatter(X, hue=y, tsne=2, title='Original data (t-SNE)')
```

```python
"""
Xt will hold values with zero-mean and unit-variance
"""
Xt, yt = YeoJohnson().fit_transform(X, y)

print('Transformed range')
print('Min', Xt.min(axis=0))
print('Max', Xt.max(axis=0))
print('Avg', Xt.mean(axis=0))
print('Std', Xt.std(axis=0))

scatter(Xt, hue=yt, title='Transformed data')
scatter(Xt, hue=yt, tsne=2, title='Transformed data (t-SNE)')
```

```python
"""
Port to C++
"""
print(YeoJohnson().fit(X, y).port(language='cpp'))
```

```python
"""
Port to JS
"""
print(YeoJohnson().fit(X, y).port(language='js'))
```
