# everywhereml.data.preprocessing.SelectKBest

See [sklearn.feature_selection.SelectKBest](https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectKBest.html)

```python
import sys
sys.path.append('../../../../')

from everywhereml.data.preprocessing.feature_selection import SelectKBest
from everywhereml.data.plot import scatter
from sklearn.datasets import load_iris
```

```python
"""
Load Iris
"""
X, y = load_iris(return_X_y=True)

print('Number of input features', X.shape[1])

scatter(X, hue=y)
scatter(X, hue=y, tsne=2)
```

```python
"""
Xt will contain the most informative features according to the score function (chi2 by default)
"""
Xt, yt = SelectKBest(k=2).fit_transform(X, y)

print('Number of output features', Xt.shape[1])

scatter(Xt, hue=yt)
scatter(Xt, hue=yt, tsne=2)
```

```python
"""
Port to C++
"""
print(SelectKBest(k=2).fit(X, y).port(language='cpp'))
```

```python
"""
Port to JS
"""
print(SelectKBest(k=2).fit(X, y).port(language='js'))
```
