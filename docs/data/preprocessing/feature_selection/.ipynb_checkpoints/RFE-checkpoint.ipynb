# everywhereml.data.preprocessing.RFE

See [sklearn.feature_selection.RFE](https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html)

```python
import sys
sys.path.append('../../../../')

from everywhereml.data.preprocessing.feature_selection import RFE
from everywhereml.data.plot import scatter
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
```

```python
"""
Load Iris
"""
X, y = load_iris(return_X_y=True)

print('Number of input features', X.shape[1])

scatter(X, hue=y)
scatter(X, hue=y, tsne=2)
```

```python
"""
Xt will contain the most informative features according to the estimator
"""
Xt, yt = RFE(k=2, estimator=RandomForestClassifier(5)).fit_transform(X, y)

print('Number of output features', Xt.shape[1])

scatter(Xt, hue=yt)
scatter(Xt, hue=yt, tsne=2)
```

```python
"""
Port to C++
"""
print(RFE(k=2, estimator=RandomForestClassifier(5)).fit(X, y).port(language='cpp'))
```
