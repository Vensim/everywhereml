# everywhereml.data.preprocessing.Normalizer

See [sklearn.preprocessing.Normalizer](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.Normalizer.html).

```python
import sys
sys.path.append('../../../')

from everywhereml.data.preprocessing import Normalizer
from everywhereml.data.plot import scatter
from sklearn.datasets import load_iris
```

```python
"""
Load Iris
"""
X, y = load_iris(return_X_y=True)

print('Original range')
print('Min', X.min())
print('Max', X.max())

scatter(X, hue=y, title='Original data')
scatter(X, hue=y, tsne=2, title='Original data (t-SNE)')
```

```python
"""
Xt will hold values scaled by the sum of the features
"""
Xt, yt = Normalizer(norm='l1').fit_transform(X, y)

print('L1 range')
print('Min', Xt.min(axis=0))
print('Max', Xt.max(axis=0))

scatter(Xt, hue=yt)
scatter(Xt, hue=yt, tsne=2)
```

```python
"""
Xt will hold values scaled by the euclidean length of the features
"""
Xt, yt = Normalizer(norm='l2').fit_transform(X, y)

print('L2 range')
print('Min', Xt.min(axis=0))
print('Max', Xt.max(axis=0))

scatter(Xt, hue=yt, title='Transformed data')
scatter(Xt, hue=yt, tsne=2, title='Transformed data (t-SNE)')
```

```python
"""
Xt will hold values scaled by the max absolute value of the features
"""
Xt, yt = Normalizer(norm='inf').fit_transform(X, y)

print('Max range')
print('Min', Xt.min(axis=0))
print('Max', Xt.max(axis=0))

scatter(Xt, hue=yt)
scatter(Xt, hue=yt, tsne=2)
```

```python
"""
Port to C++
"""
print(Normalizer(norm='l1').fit(X, y).port(language='cpp'))
```

```python
"""
Port to JS
"""
print(Normalizer(norm='l2').fit(X, y).port(language='js'))
```

```python
"""
Port to PHP
"""
print(Normalizer(norm='inf').fit(X, y).port(language='php'))
```
