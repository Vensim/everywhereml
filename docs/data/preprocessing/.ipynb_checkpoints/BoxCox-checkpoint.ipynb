# BoxCox

See [sklearn.preprocessing.PowerTransformer](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PowerTransformer.html).

See [Box-Cox Wikipedia](https://en.wikipedia.org/wiki/Power_transform#Box%E2%80%93Cox_transformation)

```python
import sys
sys.path.append('../../../')

from everywhereml.data.preprocessing import BoxCox
from everywhereml.data.plot import scatter
from sklearn.datasets import load_iris
```

```python
"""
Load Iris
"""
X, y = load_iris(return_X_y=True)

print('Original range')
print('Min', X.min(axis=0))
print('Max', X.max(axis=0))
print('Avg', X.mean(axis=0))
print('Std', X.std(axis=0))

scatter(X, hue=y, title='Original')
scatter(X, hue=y, tsne=2, title='Original (t-SNE)')
```

```python
"""
Xt will hold values with transformed according to Box-Cox formula
"""
Xt, yt = BoxCox().fit_transform(X, y)

print('Transformed range')
print('Min', Xt.min(axis=0))
print('Max', Xt.max(axis=0))
print('Avg', Xt.mean(axis=0))
print('Std', Xt.std(axis=0))

scatter(Xt, hue=yt, title='Transformed')
scatter(Xt, hue=yt, tsne=2, title='Transformed (t-SNE)')
```

```python
"""
Port to C++
"""
print(BoxCox().fit(X, y).port(language='cpp'))
```

```python
"""
Port to JS
"""
print(BoxCox().fit(X, y).port(language='js'))
```

```python
"""
Port to PHP
"""
print(BoxCox().fit(X, y).port(language='php'))
```
