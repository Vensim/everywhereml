# everywhereml.data.preprocessing.PolynomialFeatures

See [sklearn.preprocessing.PolynomialFeatures](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)

Only `degree=2` is supported.

```python
import sys
sys.path.append('../../../')

from everywhereml.data.preprocessing import PolynomialFeatures
from everywhereml.data.plot import scatter
from sklearn.datasets import load_iris
```

```python
"""
Load Iris
"""
X, y = load_iris(return_X_y=True)

print('Number of input features', X.shape[1])
print('Min', X.min(axis=0))
print('Max', X.max(axis=0))

scatter(X, hue=y)
scatter(X, hue=y, tsne=2)
```

```python
"""
Xt will contain the input features and the cross products. 
When interaction_only=False also the squared features are included (self-product).
"""
Xt, yt = PolynomialFeatures(interaction_only=False).fit_transform(X, y)

print('Number of output features', Xt.shape[1])
print('Min', Xt.min(axis=0))
print('Max', Xt.max(axis=0))

scatter(Xt, hue=yt, title='Original features')
scatter(Xt, hue=yt, tsne=2, title='Original features (t-SNE)')
```

```python
"""
Xt will contain the input features and the cross products. 
When interaction_only=True the squared features are not included.
"""
Xt, yt = PolynomialFeatures(interaction_only=True).fit_transform(X, y)

print('Number of output features', Xt.shape[1])
print('Min', Xt.min(axis=0))
print('Max', Xt.max(axis=0))

scatter(Xt, hue=yt, title='Transformed features')
scatter(Xt, hue=yt, tsne=2, title='Transfomed features (t-SNE)')
```

```python
"""
Port to C++
"""
print(PolynomialFeatures().fit(X, y).port(language='cpp'))
```

```python
"""
Port to JS
"""
print(PolynomialFeatures().fit(X, y).port(language='js'))
```
